```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df01 = pd.read_csv('C:/Users/user/Desktop/인천광역시_동구_데이터최종전처리.csv', engine = 'python')
df01.head()
```

```python
df02 = df01[['건물1', '건물2', '건물3', '행정기관', '도서관', '주차장', '생산가능인구', '충전소', '충전소_거리']]
df03 = (df02 - df02.mean())/df02.std()
df03.head()
```

```python
def weighted_sum(df, weights):
    return df[list(weights)].mul(weights).sum(axis = 1)
```

```python
weighted_sum = weighted_sum(df03, {'건물1' : 1 ,'건물2' : 2 ,'건물3' : 3 ,'행정기관' : 3 ,'도서관' : 3 ,'주차장' : 1 ,'생산가능인구' : 1 ,'충전소' : -3 ,'충전소_거리' : -3})
```

```python
weighted_sum = pd.DataFrame(weighted_sum)
weighted_sum.head()
```

```python
weighted_sum.columns=['Total']
weighted_sum.head()
```

```python
df04 = pd.concat([df03,weighted_sum],axis=1, join='inner') 
df04.head()
```

```python
#df['rank_average'] = df['age'].rank(method='average') ex) 두 명이 공동 1등이라면 둘 다 1.5등으로 처리
df04['rank'] = df04['Total'].rank(method='min', ascending = False)    # ex) 두 명이 공동 1등이라면 둘 다 1등으로 처리
#df['rank_max'] = df['age'].rank(method='max') ex) 두 명이 공동 1등이라면 둘 다 2등으로 처리
#df['rank_first'] = df['age'].rank(method='first') ex) 두 명이 공동 1등이라면 순서가 빠른 사람을 1등으로 처리
#df['rank_dense'] = df['age'].rank(method='dense') min과 동일함. 다만 순위는 항상 1씩 증가
```

```python
gid = df01['gid']
gid = pd.DataFrame(gid)
df05 = pd.concat([df04,gid],axis=1, join='inner') 
df05.head()
```
