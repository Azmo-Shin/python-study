```python
import requests
import pandas as pd
import re

client_id = "API ID"    # 제공받은 API ID
client_secret = "API SECRET"    # 제공받은 API secret

search_word = "펌프기계"    # 검색어
encode_type = "json"    # 저장형싱 xml or json
max_display = 100    # 출력 뉴스 개수
sort = 'date'    # 결과값 정렬 기준 date or sim (시간순, 관련도)
start = 1 # 출력 위치

url = f"https://openapi.naver.com/v1/search/news.{encode_type}?query=/{search_word}&display={str(int(max_display))}&start={str(int(start))}&sort={sort}"
```

```python
# 헤더에 API 키 대입
headers = {'X-Naver-Client-Id' : client_id,
           'X-Naver-Client-Secret' : client_secret
          }

# HTTP요청
r = requests.get(url, headers=headers)

# 요청결과 확인
print(r)
```

```python
r.json()
```

```python
# 데이터프레임으로 저장
df = pd.DataFrame(r.json()['items'])
```

```python
# html 정규표현식 적용
def clean_html(x):
    x = re.sub("\&\w*\;","",x)
    x = re.sub("<.*?>","",x)
    return x

df['title'] = df['title'].apply(lambda x: clean_html(x))
df['description'] = df['description'].apply(lambda x: clean_html(x))
```

```python
df.to_csv(f'new_search_result_{search_word}.csv')
```
