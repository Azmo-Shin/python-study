```python
import requests
from bs4 import BeautifulSoup
```

```python
result = []
month = ['202001','202002','202003','202004','202005','202006','202007','202008']
for i in range(len(month)):
    url ='http://openapi.airport.kr/openapi/service/AviationStatsByAirport/getTotalNumberOfFlight?\
인증키\
from_month={}&to_month={}&periodicity=0&pax_cargo=Y&domestic_foreign=I'.format(month[i],month[i])
    res = requests.get(url)
    text = res.text
    soup = BeautifulSoup(text, 'lxml-xml')
    items = soup.select('response > body > items > item')
    for item in items:
        result.append([month[i],
                      item.airport.text,
                      item.arrFlight.text,
                      item.depFlight.text,
                      item.flights.text])
```

```python
result
```

```python
import pandas as pd
```

```python
flight = pd.DataFrame(result,columns=['month','airport','arrFlight','depFlight','flights'])
```

```python
flight['month'] = flight['month'].astype('int64')
flight['arrFlight'] = flight['arrFlight'].str.replace(',', '').astype('int64')
flight['depFlight'] = flight['depFlight'].str.replace(',', '').astype('int64')
flight['flights'] = flight['flights'].str.replace(',', '').astype('int64')
```

```python
flight.to_csv('공항_별_항공통계.csv')
```
